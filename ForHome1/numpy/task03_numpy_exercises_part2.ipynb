#### Task 4: Python vs NumPy

Use pure Python and `numpy` to create a Python list/`numpy`:

- array of $10$ zeros;
- array of $10$ ones;
- array of $10$ fives;
- array of $10$ zeroes, the fifth value of which is $1$;
- array of all the integers from $10$ to $50$;
- array of all the even integers from $10$ to $50$;
- array that is the reversal of the list `[1, 2, 3]` (save these values in an object and use both methods to reverse that object);
- a 3x3 matrix with values ranging from $0$ to $8$;
- a 3x3 identity matrix;
- a 5x5 matrix with values 1,2,3,4 just below the diagonal.

````python
# Your code here
print("\nArray of 10 zeros")
py_zeros = [0 for i in range(10)]
print(py_zeros)
np_zeros = np.zeros(10)
print(np_zeros)

print("\nArray of 10 ones")
py_ones = [1 for i in range(10)]
print(py_ones)
np_ones = np.ones(10)
print(np_ones)

print("\nArray of 10 fives")
py_fives = [5 for i in range(10)]
print(py_fives)
np_fives = np.full(10, 5)
print(np_fives)

print("\nArray of 10 zeros the fifth is 1")
py_zeros_with_one = [0 for i in range(10)]
py_zeros_with_one[5] = 1
print(py_zeros_with_one)
np_zeros_with_one = np.eye(1, 10, 5)
print(np_zeros_with_one)

print("\nArray of integers from 10 to 50")
py_ten_to_fifty = [i for i in range(10, 51)]
print(py_ten_to_fifty)
np_ten_to_fifty = np.arange(10, 51)
print(np_ten_to_fifty)

print("\nArray of even integers from 10 to 50")
py_even_ten_to_fifty = [i for i in range(10, 51, 2)]
print(py_even_ten_to_fifty)
np_even_ten_to_fifty = np.arange(10, 51, 2)
print(np_even_ten_to_fifty)

print("\nReversal of [1, 2, 3]")

py_three_to_one = [1, 2, 3]
py_three_to_one.reverse()
print(py_three_to_one)
one_to_three = [1, 2, 3]
np_three_to_one = np.flip(one_to_three)
print(np_three_to_one)

print("\n3x3 Matrix from 0 to 8")
py_zero_to_eight = [[j + 3 * i for j in range(3)] for i in range(3)]
print(py_zero_to_eight)
np_zero_to_eight = np.arange(9).reshape(3, 3)
print(np_zero_to_eight)

print("\n3x3 Identity matrix")
py_identity = [[int(i == j) for j in range(3)] for i in range(3)]
print(py_identity)
np_identity = np.identity(3)
print(np_identity)

print("\nMatrix with 1 to 4 below the diagonal")
py_one_to_four_below_diag = [[(i - j == 1) * (j + 1) for j in range(5)]
                             for i in range(5)]
print(py_one_to_four_below_diag)

np_one_to_four_below_diag = np.zeros((5, 5))
for i in range(4):
    np_one_to_four_below_diag[i + 1][i] = i + 1
print(np_one_to_four_below_diag)
````
